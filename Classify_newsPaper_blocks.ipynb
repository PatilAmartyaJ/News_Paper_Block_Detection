```python
# predict.py
from ultralytics import YOLO
import cv2
import os
import numpy as np

def predict_on_image(model, image_path, output_dir="predictions", conf_threshold=0.3):
    """Make predictions on a single image"""
    os.makedirs(output_dir, exist_ok=True)

    # Run prediction
    results = model.predict(
        image_path,
        conf=conf_threshold,
        iou=0.45,
        save=False,  # We'll save manually to control output
        show=False,
        verbose=False
    )

    # Get the result
    result = results[0]
    image_name = os.path.basename(image_path)

    # Visualize results
    if result.boxes is not None or result.masks is not None:
        # Draw predictions on image
        annotated_image = result.plot()

        # Save annotated image
        output_path = os.path.join(output_dir, f"annotated_{image_name}")
        cv2.imwrite(output_path, annotated_image)
        print(f"Saved annotated image: {output_path}")

        # Print detection results
        print(f"\nResults for {image_name}:")
        if result.boxes is not None:
            print(f"Bounding boxes detected: {len(result.boxes)}")
            for i, box in enumerate(result.boxes):
                class_id = int(box.cls)
                class_name = model.names[class_id]
                confidence = float(box.conf)
                print(f"  {i+1}. {class_name}: {confidence:.3f}")

        if result.masks is not None:
            print(f"Segmentation masks detected: {len(result.masks)}")
            for i, mask in enumerate(result.masks):
                class_id = int(result.boxes[i].cls) if result.boxes is not None and i < len(result.boxes) else "Unknown"
                class_name = model.names[class_id] if isinstance(class_id, int) else class_id
                print(f"  {i+1}. {class_name} mask")

        # Save raw results (optional)
        result.save_txt(os.path.join(output_dir, f"results_{os.path.splitext(image_name)[0]}.txt"))

    else:
        print(f"No objects detected in {image_name}")

    return result

def predict_on_folder(model, folder_path, output_dir="predictions", conf_threshold=0.3):
    """Make predictions on all images in a folder"""
    supported_formats = ['.jpg', '.jpeg', '.png', '.bmp', '.tiff']
    image_files = [f for f in os.listdir(folder_path)
                  if os.path.splitext(f)[1].lower() in supported_formats]

    print(f"Found {len(image_files)} images in {folder_path}")

    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)
        predict_on_image(model, image_path, output_dir, conf_threshold)

def main():
    # Load the best trained model
    model_path = "runs/news_article_segmentation/weights/best.pt"

    if not os.path.exists(model_path):
        print("Trained model not found! Please train the model first.")
        return

    print("Loading trained model...")
    model = YOLO(model_path)

    # Create predictions directory
    os.makedirs("predictions", exist_ok=True)

    # Choose prediction mode
    print("\nChoose prediction mode:")
    print("1. Predict on a single image")
    print("2. Predict on all images in a folder")
    print("3. Use webcam for real-time detection")

    choice = input("Enter your choice (1-3): ").strip()

    if choice == "1":
        image_path = input("Enter the path to the image: ").strip()
        if os.path.exists(image_path):
            predict_on_image(model, image_path)
        else:
            print("Image not found!")

    elif choice == "2":
        folder_path = input("Enter the path to the folder: ").strip()
        if os.path.exists(folder_path):
            conf_threshold = float(input("Enter confidence threshold (0.1-0.9, default 0.3): ") or "0.3")
            predict_on_folder(model, folder_path, conf_threshold=conf_threshold)
        else:
            print("Folder not found!")

    elif choice == "3":
        # Real-time webcam detection
        print("Starting webcam detection... Press 'q' to quit.")

        cap = cv2.VideoCapture(0)
        if not cap.isOpened():
            print("Cannot open webcam!")
            return

        while True:
            ret, frame = cap.read()
            if not ret:
                print("Can't receive frame. Exiting...")
                break

            # Run prediction
            results = model.predict(frame, conf=0.3, verbose=False, imgsz=320)
            annotated_frame = results[0].plot()

            # Display results
            cv2.imshow('News Article Detector', annotated_frame)

            # Press 'q' to quit
            if cv2.waitKey(1) == ord('q'):
                break

        cap.release()
        cv2.destroyAllWindows()

    else:
        print("Invalid choice!")

    print("\nPrediction completed! Check the 'predictions' folder for results.")

if __name__ == "__main__":
    main()
```
